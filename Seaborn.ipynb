```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
weight = [62,64,69,75,66,65,71,76,73,68]
height = [120,136,148,175,137,165,154,172,200,187]
```

```python
%matplotlib inline
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight")
sns.scatterplot(x=height, y=weight)
plt.figure(figsize=(4.5,2.5))
#plt.show()-----one more method to view the scatterplot
```

```python
gender = ["female","female","female","female","female","male","male","male"]
sns.countplot(x=gender)
plt.figure(figsize=(2,1))
#plt.title("COUNT OF MALE AND FEMAES")
plt.show()
```
