```python
#H0 -->Data is normaly distributed 
#Ha ---> Data is not normay distributed

#checkikng wheather to accept the null hypothesisn take or  reject it 

sh_stat,sh_p=shapiro(wine["alcohol"])
print("shapiro statistica value : " ,sh_stat)
print("shapiro p-vaue ", sh_p)

alpha =0.05
if sh_p > alpha:
    print("Data is normally distributed sccepting h0")
else:
    print("Data is not normally distributed ,accepting Ha and rejecting H0")
    


```
