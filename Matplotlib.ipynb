```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.linspace(4,15,100)
y=np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(5,3))
plt.plot(x,y,label="Sin wave", linestyle="-.",linewidth=2,color="#234566")
plt.xlabel("X values")
plt.ylabel("Y Values")
plt.title("Sin Wave")
plt.legend()
plt.grid(True)
plt.show()
        
```

```python
category=["A","B","C","D","E"]
values=[33,54,42,51,29]
```

```python
plt.figure(figsize=(5,3))
plt.bar(category,values,alpha=0.5,edgecolor="black",color=["blue","orange","yellow","red","green"])
plt.xlabel("categories")
plt.ylabel("Values")
plt.title("Comparing categories")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x=np.random.rand(100)
y=np.random.rand(100)
color=np.random.rand(100)
size=1000*np.random.rand(100)
```

```python
sizes=np.random.randint(35,89,4)
labels=["EAST","WEST","NORTH","SOUTH"]
plt.figure(figsize=(10,3))
plt.title("Pie-chart")
plt.pie(sizes,labels=labels,colors=["lightcoral","yellow","red","cyan"],
        autopct="%1.1f%%",explode=[0,0.1,0,0])
plt.show()
```

```python
data = np.random.normal(loc=0,scale=1,size=(100,))
plt.figure(figsize=(5,3))
plt.boxplot(data, patch_artist=True, vert=False)
plt.title("Box Plot")
plt.show()
```

```python
angles=np.linspace(0,2*np.pi,100)
radii=np.random.rand(100)
```

```python
plt.figure(figsize=(5,3))
ax=plt.subplot(111,polar=True)
ax.plot(angles,radii,color="cyan",linewidth=0.2)
ax.fill(angles,radii,color="cyan",alpha=0.3)
plt.title("Polar- Plot")
plt.show()
```

```python
matrix=np.random.rand(5,5)
plt.figure(figsize=(5,3))
plt.imshow(matrix,cmap="plasma")
plt.title("Heatmap")
plt.colorbar(label="Color Sensityvity")
plt.show()
```

```python
overs=np.arange(1,21)
ind=np.random.randint(1,24,20)
nz= np.random.randint(1,24,20)
```

```python
ind_total=[0]
for i in ind:
    ind_total.append(i+ind_total[-1])

nz_total=[0]
for i in nz:
    nz_total.append(i+nz_total[-1])
```

```python
plt.subplot(121)
plt.plot(overs,nz_total[1:],label="India")
plt.plot(overs,ind_total[1:],label="New Zealand")
plt.legend()

plt.subplot(122)
plt.bar(overs,nz,label="India")
plt.bar(overs,ind,label="New Zealand")
plt.legend()
plt.show()
```
