```python
ssd= []
for i  in df["SSD"]:
    if "GB" in i :
        ssd.append(int(i.replace("GB","")))
    elif "TB" in i :
        ssd.append(int(i.replace("TB",""))*1024)
    else:
        ssd.append(int(i))
```
